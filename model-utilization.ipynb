```python
import numpy as np
import pandas as pd
import warnings
import pickle as pkl
warnings.filterwarnings('ignore')
```

```python
company = input("Enter company:")
name = input("Enter name:")
year = int(input("Enter year:"))
kms_driven = int(input("Enter kms:"))
fuel_type = input("Enter fuel type:")
```

```python
data = [[company, name, year, kms_driven, fuel_type]]
columns = ["company", "name", "year", "kms_driven", "fuel_type"]
myinput = pd.DataFrame(data = data, columns = columns)
myinput
```

```python
pipe = pkl.load(open("CPP.pkl", "rb+"))
result = pipe.predict(myinput)
print("You can purchase this car for:", round(result[0,0]))
```
